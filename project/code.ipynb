```python
df_meas = pd.read_csv("data/data_response_graphs_0.csv")
```

```python
countries = ["Sweden", "Netherlands", "United Kingdom", "Italy"]
countrydata = {}

for country in countries:
    countrydata[country] = {}
    countrydata[country]["cases"] = np.array(df_world.loc[df_world["location"] == country, "new_cases"].values)
    countrydata[country]["dates"] = np.array(df_world.loc[df_world["location"] == country, "date"].values)
    countrydata[country]["deaths"] = np.array(df_world.loc[df_world["location"] == country, "new_deaths"].values)
    countrydata[country]["measures"] = {}
    countrydata[country]["measures"]["type"] = df_meas.loc[df_meas.Country == country, "Response_measure"].values
    countrydata[country]["measures"]["start"] = df_meas.loc[df_meas.Country == country, "date_start"].values
    countrydata[country]["measures"]["end"] = df_meas.loc[df_meas.Country == country, "date_end"].values
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 12))
ax = ax.flatten()

for (i, c) in enumerate(countries):
    ax[i].plot(np.arange(len(countrydata[c]["dates"])), countrydata[c]["cases"], label=c)
    ax[i].legend()
```
